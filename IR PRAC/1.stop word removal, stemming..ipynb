```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
```

```python
# Download necessary NLTK data files (only first time)
nltk.download('punkt')
nltk.download('stopwords')
```

```python
# Input text
text = """Text preprocessing is an essential step in Natural Language Processing.
It involves cleaning and preparing text data for analysis or modeling."""
```

```python

print("Original Text:\n", text)
```

```python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')  # <-- Add this new one
nltk.download('stopwords')
```

```python
# Step 1: Tokenization
tokens = word_tokenize(text)
print("\nTokenized Words:\n", tokens)
```

```python
# Step 2: Stop word removal
stop_words = set(stopwords.words('english'))
filtered_tokens = [word for word in tokens if word.lower() not in stop_words and word.isalpha()]
print("\nAfter Stop Word Removal:\n", filtered_tokens)
```

```python
# Step 3: Stemming
ps = PorterStemmer()
stemmed_tokens = [ps.stem(word) for word in filtered_tokens]

print("\nAfter Stemming:\n", stemmed_tokens)
```

```python
# Step 4: Combine into preprocessed text
preprocessed_text = ' '.join(stemmed_tokens)
print("\nPreprocessed Text:\n", preprocessed_text)
```
